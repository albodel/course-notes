##### monocrhomatic signals
useful applications in summarizing waves, signal processing, file compression, 

Time series of a surface elevation variation $\eta(t)$
a wave with a single frequency (f=1/t)

$\eta(t)=\eta_{max}cos(2\pi ft)$

$\eta(t)=\eta_{max}cos(2\pi ft-\phi)$

$\eta(t)=\alpha cos(2\pi ft)+\beta sin(2\pi ft)$

$\eta_{max}=\sqrt{\alpha^2+\beta^2}$

##### non- monochromatic signals

$\eta(t)=\bar\eta + \eta_{max}cos(2\pi ft-\phi_1)+\eta_{max}cos(2\pi ft-\phi_2)+ \eta_{max}cos(2\pi ft-\phi_3)$

becomes

$\eta(t)=\bar\eta+\sum^3_{i=1}y_{max_i} cos(2\pi f_it-\phi_i)$

$\eta(t)=\bar\eta+\sum^3_{i=1}[\alpha_i cos(2\pi f_it)+\beta_i(2\pi f_it)]$

an infinite number of sin waves can be added to each other to make any signal in existence

this includes both repeating waveforms and non repeating waveforms

##### Practical Limitations
- In practice we cannot fit an infinite set of sines and cosines to a time series that we collect in the ocean
- With a limited set of discrete observations, only a limited set of frequencies can be fit to the data
- the lowest frequency (longest period) also called the fundamental frequency 
$f_f=\frac{1}{t_{max}}$ where $t_{max}$ =record length

$f_f=\frac{1}{N\Delta t}$ where $\Delta t$ =sampling interval and N=# of samples


##### Fourrier Analysis
model y(t) with the "harmonics" of the fundamental frequency

$f_f, 2f_f, 3f_f, 4f_f...\frac{N}{2}f_f$

or the lowest frequency we measured (the fundamental frequency) to the highest frequency (the nyquist frequency)

at each frequency $f_k$ the wave has variance:

$\sigma^2_{yk}=\frac{1}{2}(\alpha^2_k+\beta^2_k)$

Total variance ove $y(t) =\frac{1}{2}y^2_{maxk}$

$\sigma^2_{ytotal}= \frac{1}{2}\sum^{N/2}_{k=1}(\alpha^2_k+\beta^2_k)$

bigger waves = variance
