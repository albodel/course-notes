## Error propagation ##

How does error propagate through calculations and anlysis

Imagine you have three variables: <br>
a, b, c <br>
Each of those variables has a measure of uncertainty: <br>
$\sigma_a,\sigma_b,\sigma_c$

Some calculated variable $ y = f(a,b,c) $ <br>
__What is the uncertainty of y? __

### $$ \sigma_y ^2 \approx \sigma_a ^2 (\frac{dy}{da}) + \sigma_b ^2 (\frac{dy}{db}) + \sigma_c ^2 (\frac{dy}{dc}) + 2\sigma_{ab} ^2 (\frac{dy}{da})(\frac{dy}{db}) + 2\sigma_{bc} ^2 (\frac{dy}{db})(\frac{dy}{dc}) + 2\sigma_{ac} ^2 (\frac{dy}{da})(\frac{dy}{dc}) $$ ###


__Special cases__ (uncorrelated measurement variabiles):<br>
__Addition, subtractions__ : 

$ y = a + b - c $

error: $\sigma_y ^2  = \sigma_a ^2 + \sigma_b ^2 - \sigma_c ^2 $

__Multiplications, division__

$ y = \frac{ab}{c} $

$ (\frac{\sigma_y }{y})^2 =(\frac{\sigma_a }{a})^2 + (\frac{\sigma_b }{b})^2 + (\frac{\sigma_c }{c})^2  $

__ Multiplying by an exact number __

$ y=ma $

$(\frac{\sigma_y}{y})^2 = (\frac{\sigma_a}{a})^2 $

__Power__

$y = a^m$

$ (\frac{\sigma_y}{y})^2 = m (\frac{\sigma_a}{a})^2 $

__Natural log __ 

$ y= lna $

$\sigma_y = \frac{\sigma_a}{a} $

__Log 10__

$ y = log_{10}a $

$\sigma_y = 0.434 \frac{\sigma_a}{a} $

__ Exponential __

$ y = e^a $

$\sigma_y = y \sigma_a$

__ Exponential - Base 10 __

$ y = 10^a $ 

$\sigma_y = 2.303 y \sigma_a$

__Example__: Measure diameter of a sphere:

relative error = 5%

Error in calculating volume?

$V = \frac{4}{3} \pi r^3 $

$ (\frac{\sigma_{r^3}}{r^3})^2 = 3(\frac{\sigma_a}{a})^2 $

$ (\frac{\sigma_{r^3}}{r^3})^2 = 3(0.05)^2 $

$ \frac{\sigma_{r^3}}{r^3} = \sqrt{3(0.05)^2} $

$ \frac{\sigma_{r^3}}{r^3} = \sqrt{3}(0.05) $

Coefficients and phase lags can summarize difference between locations

```python
y = np.array([6,-4,27])
A = np.array([[1,1,1],
              [0,2,5],
              [2,5,-1]])
print(A.shape)
```

```python
np.linalg.solve(A,y)
```

```python
inv_A = np.linalg.inv(A)
print('Inverse Methods:', np.dot(inv_A,y))
```

```python
sst = np.array([7.6,7.4,8.2,9.2,10.2,11.5,12.4,13.4,13.7,11.8,10.1,9.0,8.9,9.5,10.6,11.4,12.9,12.7,13.9,14.2,13.5,11.4,10.9,8.1])

t = np.arange(1,25,1)
c1 = np.repeat(1,24)
c2 = np.sin((2*np.pi* t)/12)
c3 = np.cos((2*np.pi* t)/12)

A = np.array([c1,c2,c3])
A = A.transpose()


coff = np.linalg.lstsq(A,sst)[0]

print(c1.shape)
print(c2.shape)
print(c3.shape)

print(A.shape)
```

```python
fig, ax = plt.subplots()
x = np.linspace(1,24,200)

y = coff[0] + coff[1]*np.sin(2 * np.pi * x / 12) + coff[2]*np.cos(2 * np.pi * x / 12)
ax.plot(t,sst)
ax.plot(x,y)
```
